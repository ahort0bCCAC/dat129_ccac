```python
def build_crypto_url(sym, key):
    '''
    Composes an API request URL to the ...
    '''
    cryptoAPI_Endpoint = "https://financialmodelingprep.com/api/v3/historical-price-full/crypto/"
    full_url = cryptoAPI_Endpoint + sym + '?apikey='+ key
#     print("URL: ", full_url)
    return full_url
```

```python
def get_jsonparsed_data(url):
    
    # Receive the content of ``url``, parse it as JSON and return the object.
    response = urlopen(url)
    data = response.read().decode("utf-8")
    return json.loads(data)
print(get_jsonparsed_data(crypto_url))
```

```python
def make_crypto_api_req(url):
    # send the reuqest over the network and store response in resp
    response = requests.get(url)
    print("Made request, response status: ", response.status_code)
    # 200 is OK, 
    # 404 mean "I don't have that file",
    # 500-level codes are server errors
    
    #check for status code 200
    if(int(response.status_code) == 200):
        payload_objects = json.loads(response.text)
        return payload_objects
    else:
        return None
```
