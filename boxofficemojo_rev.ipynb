```python
import requests
from bs4 import BeautifulSoup
```

```python
def build_URL(tab):
    base_url = "https://www.boxofficemojo.com/"
    full_url = base_url + tab + "/"
    return full_url
```

```python
def get_url_tab():
    print("BoxofficeMojo data options")
    print("--------------------------")
    print("""
    Enter 1 for daily data.
    Enter 2 for weekend data.
    Enter 3 for weekly data.
    Enter 4 for monthly data.
    Enter 5 for quarterly data.
    Enter 6 for yearly data.
    Enter 7 for seasons data.
    Enter 8 for holidays data.          
    """)
    optSelect=input("Enter option: ")
    if optSelect=="1":
        url_tab = 'date'
    elif optSelect=="2":
        url_tab = 'weekend'
    elif optSelect=="3":
        url_tab = 'weekly'
    elif optSelect=="4":
        url_tab = 'month'
    elif optSelect=="5":
        url_tab = 'quarter'
    elif optSelect=="6":
        url_tab = 'year'
    elif optSelect=="7":
        url_tab = 'season'
    elif optSelect=="8":
        url_tab = 'holiday'
    return url_tab
```

```python
url = build_URL(get_url_tab())
print(url)
table = BeautifulSoup(requests.get(url).text, 'html.parser').find("table")
table1 = table.find_all('tr')

for i in range(1, len(table1)):
    tds = table1[i].find_all('td')
    print(i, tds)
#     print(i, tds[5].text.replace('<td class="a-text-left mojo-field-type-release mojo-cell-wide">', '')," ", 
#          tds[1].text.replace('<td class="a-text-right mojo-field-type-money">', ''))
```
